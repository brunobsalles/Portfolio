```python
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd

orders = pd.read_csv("operacoes.csv", parse_dates=True, dayfirst=True, index_col=0, decimal=",")
orders['Value'] = orders.Price * orders.Qtd
orders = orders.sort_index()
start_date = orders.index.min()
end_date = orders.index.max()

orders['Ticker'] = [s + '.SA' for s in orders['Ticker']]
tickers = orders['Ticker'].unique()

dates=pd.date_range(start_date,end_date)
portfolio=pd.DataFrame(index=dates)



```

```python
# Calculo do saldo de cada ticker
orders['Tck_Balance'] = orders.groupby('Ticker')['Qtd'].cumsum()

#Calculo do valor acumulado do portfolio
orders['Portfolio_Sum'] = orders['Value'].cumsum()

portfolio=pd.DataFrame(index=dates, columns=tickers)
```
